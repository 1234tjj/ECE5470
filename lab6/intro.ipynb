### 3. A larger 3D test image

```python
#  demonstration of visualization tools for 3D shapes
exec(vx.vxsh("  vgenim s=60 z=60 e=25,12,12 of=im1  "))
exec(vx.vxsh("  vgenim s=60 z=60 r=8,20,20 of=im2  "))
exec(vx.vxsh("  vop -or im1 im2 of=im3   "))
exec(vx.vxsh("  vdim -c im3 of=im4  "))
exec(vx.vxsh("  v3pol im4 of=im5  "))
exec(vx.vxsh("  v3pfilt -a im5 of=im6  "))
exec(vx.vxsh('  v3view if=im6 of=im7  '))
exec(vx.vxsh('  v3view -yo if=im6 ang=120 | vtile n=3,1 of=im8 '))
vd.dispmvx('im7')
vd.dispmvx('im8')
```
