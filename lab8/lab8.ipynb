### Discussion:
1. About Keras and PyTorch:
Keras abstracts away the computation backend, and we don't need to declare a lot of details when using this kind of model. PyTorch requires us to input explicit process in each stage, which gives us more flexibility but requires a longer execution time.

2. About the two models:
- Model 1 has more channels in the convolutional layers, has one more fully connected layer than Model 2, and has double number of epochs. Therefore, it is expected that Model 1 has higher computational cost than Model 2; K1 is expected to have longer running time than K2, and P1 is expected to have longer running time than P2. However, P1 runs faster than P2 maybe because I run P2 at the night of the original deadline, so a lot of people are using the server.

- Model 1 uses double number of epochs, so it is expected that the training and testing accuracy to be higher for Model 1. It is true for the Keras version, while the accuracy for the PyTorch version is similar. Maybe it's because the accuracy for the PyTorch model is already very high (more than 98%).

- They use different approaches to prevent the over-fitting problem. Model 1 has two rounds of dropouts, which introduces some randomness in training. Model 2 uses batch normalization to reduce the effect of strange or extreme values.

- Model 1 is more suitable for complex tasks and larger datasets, while Model 2 is more suitable for simpler tasks and smaller datasets.
